```python
import pandas as pd

df_lang = pd.read_csv("popular-languages.csv")
df_lang.head()
```

```python
df_lang.columns
```

```python
df_lang['Average Salary'] = df_lang['Average Salary'].replace('[\$,]', '', regex=True).astype(float)
```

```python
df_lang_sorted = df_lang.sort_values(by='Average Salary', ascending=False)
df_lang_sorted
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.bar(df_lang_sorted['Language'], df_lang_sorted['Average Salary'])
plt.xticks(rotation=45)
plt.xlabel("Programming Language")
plt.ylabel("Average Salary")
plt.title("Popular Programming Languages by Salary")
plt.tight_layout()
plt.show()
```

```python
import requests
import pandas as pd

api_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/module%201/Accessing%20Data%20Using%20APIs/jobs.json"

response = requests.get(api_url)
data = response.json()

df_jobs_api = pd.DataFrame(data)

df_jobs_api.columns
```

```python
import re

def get_number_of_jobs_T(technology):
    pattern = r'\b' + re.escape(technology) + r'\b'
    filtered = df_jobs_api[df_jobs_api["Key Skills"].str.contains(
        pattern, case=False, na=False, regex=True)]
    return technology, filtered.shape[0]
```

```python
technologies = [
    "C",
    "C#",
    "C++",
    "Java",
    "JavaScript",
    "Python",
    "Scala",
    "Oracle",
    "SQL Server",
    "MySQL",
    "PostgreSQL",
    "MongoDB"
]

results = []

for tech in technologies:
    tech_name, count = get_number_of_jobs_T(tech)
    results.append([tech_name, count])

results_df = pd.DataFrame(results, columns=["Technology", "Number of Job Postings"])

results_df
```

```python
results_df.to_excel("job-postings.xlsx", index=False)
```
